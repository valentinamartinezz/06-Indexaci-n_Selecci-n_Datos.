```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ventas = pd.read_csv('Data/ventas.csv',encoding ='latin-1')
```

## 1.Acceso a columnas y estadísticas básicas:

#### Imprime las primeras 5 filas del DataFrame.

```python
ventas.head(5)
```

#### Calcula la media y la desviación estándar de la columna num_passengers. (Aplicar método mean() y std())

```python
media_ventas = ventas['num_passengers'].mean()
print(f"Media de vuelos: {media_ventas}")
```

```python

desviacion_ventas = ventas['num_passengers'].std()
print(f"Desviación estándar de vuelos: {desviacion_ventas}")
```

#### ¿Cuál es el número máximo de pasajeros registrados? (Aplicar método max()).


```python
max_ventas = ventas['num_passengers'].max()
print(f"Máximo de vuelos: {max_ventas}")
```

## 4.Agregación y agrupación:

#### Agrupa los datos por ruta y calcula la suma total de pasajeros para cada ruta. (aplicar groupby(), sum()).

```python
Ruta=ventas.groupby('route')['num_passengers'].sum().sort_values(ascending=True)
```

## 6.Visualización de datos:

#### Crea un gráfico de barras que muestre la cantidad de pasajeros por día de la semana (flight_day).

```python
demanda_semanal=ventas.groupby('flight_day')['num_passengers'].mean()
```

```python
demanda_semanal.plot.bar(x=demanda_semanal.index,y=demanda_semanal,rot=0)
plt.title('Demanda por dia de la semana')
plt.xlabel('Dia de la semana')
plt.ylabel('Numero de pasajeros promedio')
plt.show()
```
